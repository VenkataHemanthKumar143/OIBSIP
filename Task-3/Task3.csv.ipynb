```python
q1=ds["price"].quantile(0.25)
q3=ds["price"].quantile(0.75)
IQR=q3-q1
IQR
```

```python
upper_limit=q3+1.5*IQR
lower_limit=q1-1.5+IQR
upper_limit,lower_limit
```

```python
def imputer(value):
    if value>upper_limit:
        return upper_limit
    if value<lower_limit:
        return lower_limit
    else:
        return value
```

```python
ds["price"]=ds["price"].apply(imputer)
```

```python
ds.info()
```

```python
predict = "price"
ds = ds[["symboling", "wheelbase", "carlength",
         "enginesize", "boreratio", "stroke",
         "compressionratio", "horsepower", "peakrpm",
         "carwidth", "carheight", "curbweight",
         "citympg", "highwaympg", "price"]]

a= np.array (ds.drop([predict], 1))
b= np.array (ds [predict])

from sklearn.model_selection import train_test_split
atrain, atest, btrain, btest = train_test_split(a, b, test_size=0.2)

from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(atrain, btrain) 
predictions = model.predict (atest)

from sklearn.metrics import mean_absolute_error
model.score(atest, predictions)
```
